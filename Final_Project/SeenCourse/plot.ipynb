```python
import json
import numpy as np
import matplotlib.pyplot as plt

with open('output/records_basic_32_12_layer.json', 'r') as f:
    records = json.load(f)

plt.clf()
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
plot_loss = ax1.plot(np.arange(len(records['train_loss'])), records['train_loss'], label='train_loss', color='m')
plot_mapk = ax2.plot(np.arange(len(records['eval_mapk'])), records['eval_mapk'], label='eval_mapk', color='c')

plots = plot_mapk + plot_loss
ax1.legend(plots, [l.get_label() for l in plots], loc='right')

ax1.set_xlabel('epoch')
ax1.set_ylabel('loss', color='m')
ax2.set_ylabel('mapk', color='c')
plt.show()
```
